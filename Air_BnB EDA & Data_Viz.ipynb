```python
# Univariate Analysis

# Price Distribution

sns.histplot(data=df, x='price')
```

```python
# So here in the statical measure result we can see one value is in 100000, but in the above graph its not dosplayed

# Hence checking the outlayers in the price

sns.boxplot(data=df, x='price')
```

```python
# Here the Maximum Price are <15000, one for two price are alone in 100000
#Hence ignoring the values
#Now assigning these <16000 price to new data frame ('data') for analysis

data =df[df['price']<16000]
```

```python
sns.boxplot(data=data,x='price')
```

```python
# Now checking again the price distribution

# Set the figure size
plt.figure(figsize=(8, 5))

# Create the histogram
sns.histplot(data=data, x='price',bins=120)

# Naming the chart
plt.title("Price Distibution")
```

```python
# Univariate Analysis

# availability_365

plt.figure(figsize=(8,5))

sns.histplot(data=df, x='availability_365',bins=30)

plt.title("Availability_365 Distribution")
```

```python
# Group by "neighbourhood_group" with "Price"

data.groupby('neighbourhood_group')['price'].mean()
```

```python
# Feature Engineering

#Price per Bed

data['price_per_bed']=data['price']/data['beds']
```

```python
data.head()
```

```python
# Now grouping "neighbourhood_group" with "Price_Per_Bed"

data.groupby('neighbourhood_group')['price_per_bed'].mean()
```

```python
# Bi variable Analysis (One variabel dependency in another variable)

data.columns
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=data,x='neighbourhood_group',y='price',hue='room_type')
```

```python
# Checking is there any relationship between the 'Number of Reviews' Vs 'Price'

sns.set_style('darkgrid')
plt.figure(figsize=(10,5))
plt.title("Locality and Review Analysis")
sns.scatterplot(data=data,x='number_of_reviews',y='price',hue='neighbourhood_group')
```
